## **Median using fill null values**

```python
pf['Age']=pf['Age'].fillna(pf['Age'].median())
pf['Purchase_Amount']=pf['Purchase_Amount'].fillna(pf['Purchase_Amount'].median())
pf['Feedback_Score']=pf['Feedback_Score'].fillna(pf['Feedback_Score'].median())
```

## **Mode using fill null values**

```python
pf['City']=pf['City'].fillna(pf['City'].mode()[0])
pf['Gender']=pf['Gender'].fillna(pf['Gender'].mode()[0])
pf['Subscribed']=pf['Subscribed'].fillna(pf['Subscribed'].mode()[0])
```

```python
print(pf.isnull().sum())
```

## **Fill the empty Name column in Unknown**

```python
pf['Name']=pf['Name'].fillna('Unknown')
```

```python
pf.isnull().sum()
```

```python
pf['Gender']=pf['Gender'].replace({'M':'Male','male':'Male','m':'Male','F':'Female','f':'Female','female':'Female'})
```

```python
pf['Subscribed']=pf['Subscribed'].replace({'N':'No','no':'No','n':'No','Y':'Yes','y':'Yes','yes':'Yes'})
```

```python
pf['City']=pf['City'].replace(' ' ,np.nan)
```

## **IQR (Interquartile Range) Formula**
### **Short Explanation:**
✅ q1 → 25th percentile
✅ q3 → 75th percentile
✅ IQR → Range between Q3 and Q1 (middle 50% of data)
✅ Outlier bounds:
Lower Bound = Q1 - 1.5 * IQR  
Upper Bound = Q3 + 1.5 * IQR  


```python
q1 = pf['Age'].quantile(0.25)
q3 = pf['Age'].quantile(0.75)
iqr = q3 - q1

# Filter data to remove outliers
pf = pf[(pf['Age'] >= q1 - 1.5 * iqr) & (pf['Age'] <= q3 + 1.5 * iqr)]
```

## **Drop the Subscribed Column**

```python
pf.drop('Subscribed', axis=1,inplace=True)
```
